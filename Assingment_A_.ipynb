```python
# Data structures for storing the graph
coordinates = {}
adjacency_list = defaultdict(list)

# Read coordinates from the CSV file
with open(coordinate_file, 'r') as file:
    reader = csv.reader(file)
    next(reader)  # Skip the header row
    for star_name, x, y, z in reader:
        coordinates[star_name] = (int(x), int(y), int(z))

with open(distance_file, "r") as file:
    reader = csv.reader(file)
    for source, destination, dist in reader:
        adjacency_list[source].append((destination, int(dist)))

print("Coordinates and adjacency list loaded successfully!")
print(f"Sample Coordinates: {list(coordinates.items())[:5]}")
print(f"Sample Adjacency List: {list(adjacency_list.items())[:5]}")
```

```python
def euclidean_distance(star1, star2):
    x1, y1, z1 = coordinates[star1]
    x2, y2, z2 = coordinates[star2]
    print(f"Coordinates of {star1}: (x1={x1}, y1={y1}, z1={z1})")
    print(f"Coordinates of {star2}: (x2={x2}, y2={y2}, z2={z2})")
    dx = x2 - x1
    dy = y2 - y1
    dz = z2 - z1
    print(f"Differences: dx={dx}, dy={dy}, dz={dz}")

    distance = math.sqrt(dx**2 + dy**2 + dz**2)
    print(f"Euclidean Distance between {star1} and {star2}: {distance}\n")

    return distance
```

```python
#  import heapq

def dijkstra(start, goal):
    pq = [(0, start, [])]
    visited = set()

    while pq:
        cost, current, path = heapq.heappop(pq)
        if current in visited:
            continue
        visited.add(current)
        path = path + [current]

        if current == goal:
            return cost, path

        for neighbor, weight in adjacency_list[current]:
            if neighbor not in visited:
                heapq.heappush(pq, (cost + weight, neighbor, path))

    return float("inf"), []
```

```python
def a_star(start, goal):
    pq = [(0, 0, start, [])]
    visited = set()

    while pq:
        f, g, current, path = heapq.heappop(pq)
        if current in visited:
            continue
        visited.add(current)
        path = path + [current]

        if current == goal:
            return g, path

        for neighbor, weight in adjacency_list[current]:
            if neighbor not in visited:
                g_score = g + weight
                f_score = g_score + euclidean_distance(neighbor, goal)
                heapq.heappush(pq, (f_score, g_score, neighbor, path))

    return float("inf"), []
```

```python

start_star = "Sun"
goal_star = "Upsilon Andromedae"
start_time = time.time()
dijkstra_cost, dijkstra_path = dijkstra(start_star, goal_star)
dijkstra_time = time.time() - start_time
start_time = time.time()
a_star_cost, a_star_path = a_star(start_star, goal_star)
a_star_time = time.time() - start_time
print("Dijkstra's Algorithm:")
print(f"Path: {dijkstra_path}")
print(f"Cost: {dijkstra_cost}")
print(f"Time: {dijkstra_time:.4f} seconds\n")

print("A* Algorithm:")
print(f"Path: {a_star_path}")
print(f"Cost: {a_star_cost}")
print(f"Time: {a_star_time:.4f} seconds")
```
